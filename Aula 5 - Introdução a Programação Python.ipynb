```python
a = 1 #Class Int (inteiro)
b = 1.2 #Class Float (Flotoante)
c = "a" #Class Str (String)
d = True #Class Bool (Booleano)

print(a,type(a)) 
print(b,type(b))
print(c,type(c))
print(d,type(d))

print(a+d)
print(a+b-d)
print(a+c) 

print("Int, Float e Bool: São classes númericas! Não pode misturar classes para calculos, Ex:a+c, retorna erro.")
```

Slice <- Fatia o resultado e retorna somente o que foi definido nos parâmetros.

```python
a = "Falcão"
print(a[-3:]) # Cão
print(a[3]) # C -> Somente a Letra
print(a[-3:-5:-1]) # Cl
print(a[::-1]) # Retorna inverso
print(a[2].upper()) # L -> Retorna a Letra em caixa alta
```

```python
a = ["a","b","c"]
b = ["d","e"]
print(a+b) # Retorna a, b, c, d, e
print(a + [b[0]]) # Retorna a, b, c, d
print(a+ [b[1]]) # Retorna a, b, c, e
print(a[1:3]) # Retorna b, e
```

Conjuntos!

### Não permite elementos duplicados.

```python
a = {1,2,3,4,5,6,7}
b = {6,7,8,9,10,11,12}
print("Conjunto de elementos de A:")
print(a)
print("")
print("Conjunto de elementos de B: ")
print(b)
print("")
print("União de elementos de A e B com conectivo ou: ")
print(a | b) 
print("")
print(a.union(b))
print("")
print("Interseção dos elementos que estão em A e B: ")
print(a.intersection(b))
print("Elementos dos conjuntos de A e B que são diferentes: ")
print(a-b)
print(a.difference(b))
print("")
print("Elementos dos conjuntos de B e A que são diferentes: ")
print(b-a)
print(b.difference(a))
```

### Dicionários!

```python
#Dicionários - Chave:Valor
#Simbologia: {}

imdb = {"Formula da água": 8.8,
        "Amelie Polan":9.0,
        "Sky": 5.3,
        "Venom":7.6}

imdb["Sky"] = 9.1
print(imdb)

del(imdb["Sky"])
print(imdb)

for i in imdb:
    print(i, end=" ")
    print("")
for i in imdb.values():
    print(i, end=" ")
    print("")
    print(imdb.keys())
    print("")
    print(imdb.values())
```
