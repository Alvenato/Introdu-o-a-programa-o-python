### O programa cria uma lista vazia de temperaturas e entra em um loop infinito para solicitar dados do usuário.
### Critério de parada: O loop só termina se você digitar exatamente 999. Esse é ou "valor sentinela".
### Utilização eval(): Esta função converte o texto digitado em um número (inteiro ou real).

```python
temperaturas = []
while True:
    temperaturaInfo = eval(input("Informe a temperatura: "))
    if temperaturaInfo == 999:
        break
    temperaturas.append(temperaturaInfo)
print("Temperaturas informadas: ", temperaturas)
print("Temperatura média: ", sum(temperaturas) / len(temperaturas))
print("Temperatura máxima: ", max(temperaturas))
print("Temperatura mínima: ", min(temperaturas))
if len(temperaturas) > 0:
    print("Temperatura mais próxima da média: ", min(temperaturas, key=lambda x: abs(x - (sum(temperaturas) / len(temperaturas)))))
else:
    print("Nenhuma temperatura informada.")
    
```
