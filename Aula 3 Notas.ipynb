Esta atividade consiste em um sistema básico de gerenciamento de notas escolares desenvolvido em Python. O código utiliza estruturas de programação fundamentais para coletar dados, realizar cálculos e tomar decisões com base em condições predefinidas.

```python
turma = []
while True:
    nome = input("Digite o nome do aluno (ou 'sair' para encerrar): ")
    if nome.lower() == 'sair':
        break
    turma.append(nome)
    nota1 = float(input("Digite a primeira nota do aluno: "))
    nota2 = float(input("Digite a segunda nota do aluno: "))
    media = (nota1 + nota2) / 2
    turma.append((nome, nota1, nota2, media))
    print(f"Aluno: {nome}, Nota 1: {nota1}, Nota 2: {nota2}, Média: {media:.2f}")
    if media >= 7:
        print("Situação: Aprovado")
    elif media >= 5:
        print("Situação: Recuperação")
    else:
        print("Situação: Reprovado")
```
