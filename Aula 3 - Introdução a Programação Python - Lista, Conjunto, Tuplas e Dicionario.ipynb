### Tuplas

```python
tupla = (1,2,3,4,2,1,4)
print(tupla)
#tupla[2]= 5 <- Não é possível
tupla = tupla + (6,9,10)
print(tupla)
```

```python
lista = list(tupla)
tupla2 = tuple(lista)
print(lista)
```

### Conjuntos

```python
Rgs = {1,2,3,4,2,5,6,1,4,3}
print(Rgs)

lista = list(Rgs)
print(lista)

cnj = {1,4,6,5,7,8,6}
lista = list(cnj)
tupla = tuple(lista)
cnj = set(tupla)
cnj = set(lista)
print(cnj)
```

### Dicionários

```python
i = 10
print(type(i))
i = float(i)
print(type(i))
i = str(i)
print(type(i))
i = bool(i)
print(type(i))
i = None
print(type(i))

```

### Lista -> Mutável e ordenada. Ótima para sequências que mudam.[]
### Tupla -> Imutável. Protege os dados contra alterações (mais rapidamente).()
### Conjunto -> Elementos únicos. Não utilize peças duplicadas e não as encomende.{}
### Dicionário -> Chave-valor. Busca rápida por meio de uma chave única.{}

# Solicite ao usuario varias frutas, depois armazene em uma estrutura de seu interesse.
### Quantas dessas frutas são, Morangos, Abacaxis e Bananas.

```python
frutas = []
while True:
    fruta = input("Digite o nome das frutas: ")
    if fruta == 'sair':
        break
    frutas.append(fruta)

print("Morangos: ", frutas.count("Morango"))
print("Bananas: ", frutas.count("Banana"))
print("Abacaxi: ", frutas.count("Abacaxi"))


```
