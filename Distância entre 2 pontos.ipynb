Este código é um exemplo clássico de geometria analítica aplicada à programação . Ele calcula a distância entre dois pontos em um plano bidimensional (2D) para determinar se existe risco de colisão ou proximidade excessiva.

#### Criar um objeto A (x1, y1)
#### Criar um objeto B (x2, y2)
Calcular a formula da distância, se distância for menor que 5 mostrar "Alerta de proximidade!" senão, distancia for maior que 5 mostrar "Áeea segura!"

```python
objetoA = (12, 5)
objetoB = (13, 5)
distanciaA = (objetoB[0] - objetoA[0])**2
distanciaB = (objetoB[1] - objetoA[1])**2
raiz = (1/2)
distancia = (distanciaA + distanciaB)**raiz
print("Posição do objeto A: ", objetoA)
print("Posição do objeto B: ", objetoB)
print("Distancia do objeto A para o objeto B: ", distancia)
if distancia < 5:
    print("Alerta de proximidade.")
else:
    print("Área segura!")
```
